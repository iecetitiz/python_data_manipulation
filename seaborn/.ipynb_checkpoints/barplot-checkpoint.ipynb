```python
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.shape
```

```python
sns.set_style("darkgrid")
```

```python
sns.barplot(x = "species", y = "body_mass_g", data = penguins); #x and y are the column names
```

```python
penguins.groupby("species").body_mass_g.mean() #thats why barplot is a categorical estimate plot
```

```python
#barplot kategorik degiskenlerin her birini grupluyor, grupby olarak calisiyor ve her bir sayisal degerin meanini heaplayip bar grafik olarak getiriyor
```

```python
#horizontal barplot
```

```python
sns.barplot(x = "body_mass_g", y = "species", data = penguins) #we switch x and y
```

```python
#confidence interval diye bir sey var ama simdilik geciyorum cok anlamlandiramadim
```

```python
#estimator
```

```python
sns.barplot(x = "species", y = "body_mass_g", 
            data = penguins,
            estimator = np.std
           ); 
```

```python
#lets check
penguins.groupby("species").body_mass_g.describe()
```

```python
sns.barplot(x = "species", y = "body_mass_g", 
            data = penguins,
            estimator = np.max
           ); 
```

```python
sns.barplot(x = "species", y = "body_mass_g", 
            data = penguins,
            hue = "sex"
           ); 
```

```python
#horizontal version
sns.barplot(x = "body_mass_g", y = "species", 
            data = penguins,
            hue = "sex"
           ); 
```

```python
#istedigim sirada yazabilirim
sns.barplot(x = "species", y = "body_mass_g", 
            data = penguins,
           order=["Gentoo", "Chinstrap", "Adelie"]
           ); 
```

```python
#istersem female once olsun diyebilirim
sns.barplot(x = "species", y = "body_mass_g", 
            hue = "sex",
            order=["Gentoo", "Chinstrap", "Adelie"],
            hue_order = ["Female", "Male"],
            data = penguins,
            
           ); 
```

```python
#styling colors
```

```python
#hepsi ayni renk olsun isteyebilirim
sns.barplot(x = "species", y = "body_mass_g", 
            data = penguins,
           color = "c"
           ); 
```

```python
#errcolor and errwidth
```

```python
sns.barplot(x = "species", y = "body_mass_g", 
            data = penguins,
           color = "c",
            errcolor = "darkslateblue",
            errwidth= 8
           ); 
```

```python
#palette
```

```python
#it's actually 170 palettes!!!
sns.barplot(x = "species", y = "body_mass_g", 
            data = penguins,
          palette = "hot"
           ); 
```

```python
sns.barplot(x = "species", y = "body_mass_g", 
            data = penguins,
            ci = None, 
            color = "aliceblue",
            edgecolor = "mediumblue",
            lw = 3
           ); 
```
