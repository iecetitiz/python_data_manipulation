```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

# distplot()

```python
df = sns.load_dataset("tips")
```

```python
df
```

```python
sns.histplot(df["tip"])
```

```python
sns.histplot(df["size"])
```

```python
sns.histplot(df["tip"])
```

```python
sns.histplot(df["tip"], bins = 50)
```

```python
sns.set(style = "darkgrid", font_scale = 1.5)
sns.distplot(df["total_bill"], bins = 50, color = "green", hist = False, kde_kws = {"shade": True}, rug = True)
#kde = True cizgiyi gosteriyor
```

```python
sns.set(style = "whitegrid", font_scale = 1.5)
fig, axes = plt.subplots(2, 2, figsize = (10, 10))
sns.distplot(df["size"], ax = axes[0,0]) # ax ile belirttigim yer grafigin konumu olacak
sns.distplot(df["total_bill"], bins = 50, kde = False, color = "blue", ax = axes[0,1])
sns.distplot(df["total_bill"], bins = 50, kde = False, color = "red", ax = axes[1,0])
sns.distplot(df["total_bill"], bins = 50, color = "green", hist = False, kde_kws = {"shade": True}, rug = True, ax = axes[1,1])
plt.tight_layout()
```
